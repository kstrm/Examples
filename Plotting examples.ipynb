# Plotting examples

Below are a few example to help you make publication quality figures. See notebook [02 Plotting with Matplotlib.ipynb](https://github.com/kstrm/Starting-out-with-python/blob/master/02%20Plotting%20with%20Matplotlib.ipynb) for more on plotting.

## Load packages

```python
%pylab inline
%config InlineBackend.figure_format='retina'
```

## Basic plotting

```python
# create some data

x = np.linspace(0, 5, 100)
y = x ** 2

# plot the data

fig, ax = plt.subplots() # no specific size given

ax.plot(x, x**2, color='C0',label='$y = x^{2}$')
ax.plot(x, x**3, color='C3', label='$y = x^{3}$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend(loc=2); # more examples on how to controll the legend are given below
```

## Publication plotting

### Example 1

```python
# the same data as the "basic plotting" example above, but with a specified figure size and font size

fig, ax = plt.subplots(figsize=(4,3)) # 4 in  x 3 in figure

ax.plot(x, x**2, label='$y = x^{2}$')
ax.plot(x, x**3, label='$y = x^{3}$')

ax.set_xlabel('$x$', fontsize=10)
ax.set_ylabel('$y$', fontsize=10)

ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

ax.legend(loc=2, fontsize=8); # more examples on how to controll the legend are given below

# save the figure
fig.tight_layout();
savefig('example1.pdf',bbox_inches="tight", pad_inches=0.005)
```

### Example 2

```python
# Define the data -----------

# gravel bed
ds_gr_mm = array([9.525,6.35,4.75,3.35,2.36,1.8,0.85])
perc_pass_gr = array([99.30,98.39,72.74,15.56,2.73,0.08,0.00])

# sand 
ds_sand_mm = array([1.7,1.4,1,0.71,0.5,0.3,0.25,0.18,0.15,0.075])
perc_pass_sand = array([99.25,98.08,92.21,66.23,42.64,21.41,14.34,6.71,3.73,0.64])

# sand-gravel bed 
ds_sdgr_mm = array([6.3,4.75,3.35,2.36,1.7,1.4,1,0.71,0.5,0.3,0.25,0.18,0.15,0.075])
perc_pass_sdgr = array([99.95,92.75,52.01,28.97,22.73,21.60,19.88,19.36,12.57,6.32,4.05,1.85,1.23,0.15])

# plot and save the figure -----------

fig, ax = plt.subplots(figsize=(4,3)) # size in inches

ax.plot(ds_gr_mm, perc_pass_gr, 's-', color='C1', label="Gravel")
ax.plot(ds_sand_mm, perc_pass_sand, 'o-', color='C7', label="Sand")
ax.plot(ds_sdgr_mm, perc_pass_sdgr, 'v-',color='C2', label="Sand-Gravel")

ax.set_xlabel(r'$d_{s}$ [mm]', fontsize=10)
ax.set_ylabel('Percent Passing By Weight [%]', fontsize=10)

ax.semilogx()

ax.axvline(x=2,color='k', linestyle='--', alpha=0.5)
ax.text(1.3,65, "Sand/\nGravel\nBreak", backgroundcolor='white', 
        clip_on='True', multialignment='center', alpha=0.75) #weight = 'bold', 

ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
ax.legend(loc=2, fontsize=8); # more examples on how to controll the legend are given below

fig.tight_layout();
savefig('example2.pdf',bbox_inches="tight", pad_inches=0.005)
```
